# Business cycles across countries


```python
import numpy as np
import eurostat
import matplotlib.pyplot as plt
import pandas as pd
import fredapi
import statsmodels.api as sm

CB =  ['#000000','#004488','#BB5566','#DDAA33', 'purple']
LS = ['solid', 'dashed', ':', '-.']
```

```python
# Function to generate AR(1) time series
ll = 1000
def generate_ar1_series(persistence, length=100):
    ar1_process = [0]  # Initial value

    for _ in range(1, length):
        ar1_process.append(persistence * ar1_process[-1] + np.random.normal())

    return np.array(ar1_process)

# Set different persistence values to visualize
persistence_values = [0, 1, 0.9]

# Function to plot AR(1) time series
i=0
for persistence in persistence_values:
    ar1_series = generate_ar1_series(persistence, 1000)
    plt.plot(ar1_series, label=f'Persistence = {persistence:.2f}', color = CB[i+1], linestyle = LS[i])
    plt.legend(frameon=False)
    plt.grid(color = 'grey', linestyle = '--', linewidth = 0.5)
    i+=1

plt.xlabel('Time')
plt.ylabel('Value')
plt.savefig('ar1.pdf')
plt.show()


```
